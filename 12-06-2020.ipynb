### Linear algebra

```python
from scipy import linalg
#create 2d matrix'
ar2 = np.array([[4,5],[3,2]])
linalg.det(ar2)
```

```python
linalg.inv(ar2)
```

```python
from scipy import misc
data = misc.face()
plt.imshow(data)
```
